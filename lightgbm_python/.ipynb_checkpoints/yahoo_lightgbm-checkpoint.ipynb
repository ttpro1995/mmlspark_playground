```python
parent_folder = "/data/public/dataset/Learning_to_Rank_Challenge"
```

```python
yahoo_train_path = os.path.join(parent_folder, "yahoo.train")
yahoo_test_path = os.path.join(parent_folder, "yahoo.test")
```

```python
# train_data_sklearn = load_svmlight_file(yahoo_train_path)
```

```python
# val_data_sklearn = load_svmlight_file(yahoo_validation_path)
```

```python
train_data = lgb.Dataset(yahoo_train_path)
```

```python
validation_data = lgb.Dataset(yahoo_test_path, reference=train_data)
```

```python
# learning_rate = 0.1
# num_leaves = 255
# num_trees = 500
# num_threads = 16
# min_data_in_leaf = 0
# min_sum_hessian_in_leaf = 100
```

objective=lambdarank metric=ndcg

```python
param={"objective":"lambdarank","metric":"ndcg","learning_rate":0.1, "num_leaves":255, "num_trees":500,"num_threads":4,"min_data_in_leaf":0,"min_sum_hessian_in_leaf":100}
```

```python
num_round = 1
bst = lgb.train(param, train_data, num_round, valid_sets=[validation_data])
```
